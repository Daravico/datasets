```python
import cv2

cap = cv2.VideoCapture(0)

count = 0
limit = 300

gestures = {
    'empty': 0,
    'one': 0,
    'two': 0,
    'palm': 0
}

id_gesture = 0

gesture_names = list(gestures.keys())
limit_gestures = len(gestures.keys())

# ROI:
x = 20 #int(cap.get(cv2.CAP_PROP_FRAME_WIDTH) // 4)
y = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT) // 4)

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH) // 4)
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT) // 2)


while True:

    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    k = cv2.waitKey(10)

    if k == ord('q'):
        break


    # ----------------------------------
    if k == ord('t') and count < limit:
        count += 1
        cv2.imwrite(f'data/Gesture_{str(id_gesture)}_{str(count)}.jpg', gray[y:y+h, x:x+w])
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), 2)

    elif k == ord('t') and count >= limit:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 2)

    elif k == ord('n') and count >= limit:
        count = 0
        id_gesture += 1

        if id_gesture >= limit_gestures:
            break

    elif k == -1 and count >= limit:
        cv2.putText(frame, 
                f'Press "n" to continue...', 
                (400, 150), cv2.FONT_ITALIC, 
                0.75, (255,0,0), 1)

    else:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,0), 2)

    # ----------------------------------

    cv2.putText(frame, 
                f'Gesture: {gesture_names[id_gesture]}', 
                (400, 50), cv2.FONT_ITALIC, 
                0.75, (255,0,0), 1)
    
    cv2.putText(frame, 
                f'Count: {count}', 
                (400, 100), cv2.FONT_ITALIC, 
                0.75, (255,0,0), 1)

    cv2.imshow("Frame", frame)

cap.release()
cv2.destroyAllWindows()
```
